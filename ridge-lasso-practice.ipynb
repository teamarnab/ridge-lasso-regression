```python
#Splitting data into independent and dependent variables
X = hitters.drop(['Salary'], axis=1)
X.columns
```

```python
y = hitters['Salary']
y
```

```python
#Scaling data with standard scaler
from sklearn.preprocessing import StandardScaler
scaled_data = StandardScaler()
X = scaled_data.fit_transform(X)
```

```python
X
```

```python
#Splitting into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
#Predicting using test set
test_model_predicted_val = lin_model.predict(X_test)
```

```python
#Checking accuracy
from sklearn.metrics import r2_score
test_model_score = r2_score(y_test, test_model_predicted_val)
print("The model's score is", test_model_score)
print("Accuracy %", test_model_score*100, "%")
```

```python
#Checking accuracy with training data

#Predicting with training dataset
train_model_predicted_val = lin_model.predict(X_train)

#Checking accuracy
from sklearn.metrics import r2_score
train_model_score = r2_score(y_train, train_model_predicted_val)
print("The model's score is", train_model_score)
print("Accuracy :", train_model_score*100, "%")
```

```python
#Checking accuracy with complete data

#Predicting with training dataset
total_predicted_val = lin_model.predict(X)
#Checking accuracy
from sklearn.metrics import r2_score
total_model_score = r2_score(y, total_predicted_val)
print("The model's score is", total_model_score)
print("Accuracy :", total_model_score*100, "%")
```

Accuracy with test dataset is 27%, training dataset is 58% and complete dataset is 52%.
Hence, this is an overfitted model

```python
#Creating an alpha list for Ridge
alphas = np.arange(200, 300, 10)
print(alphas)
```

Tried with different range 

from 0 to 1 with step size 1

from 10 to 100 with step size 10

from 100 to 500 with step size 100

from 200 to 300 with step size 10 gave the best accuracy

```python
#Importing ridge and fitting our data and checking the accuracy for each alpha value
from sklearn.linear_model import Ridge
for alpha in alphas:
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X_train, y_train)
    pred_value = ridge_model.predict(X_test)
    accuracy = r2_score(y_test, pred_value)
    print(f"For alpha value {alpha}, accuracy is {accuracy*100}")
```

Applying Lasso

```python
#Creating an alpha list for Ridge
alphas_lasso = np.arange(10, 100, 10)
print(alphas_lasso)
```

Tried with different range 

from 0 to 1 with step size 1

from 10 to 100 with step size 10

from 100 to 500 with step size 100

from 10 to 100 with step size 10 10 gave the best accuracy

```python
from sklearn.linear_model import Lasso
for alpha in alphas_lasso:
    lasso_model = Lasso(alpha=alpha)
    lasso_model.fit(X_train, y_train)
    pred_value = lasso_model.predict(X_test)
    accuracy = r2_score(y_test, pred_value)
    print(f"For alpha value {alpha}, accuracy is {accuracy*100}")
```
